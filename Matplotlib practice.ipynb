```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
linear = np.arange(8)
y1 = np.arange(8)
y2 = np.square(linear)
```

```python
plt.figure()
ax1 = plt.subplot(121)
plt.plot(linear,y1,'-om')
ax = plt.subplot(122,sharey =ax1)
ax.plot(linear,y2,':xy')
```

```python
ax1.plot(linear,y2,'-.')
```

```python
ax1.clear()
```

```python
ax1.fill_between(x = linear,y1 = y1,y2 = y2,where = y2>=y1)
```

```python
plt.figure()
plt.plot(linear,linear,'-om',alpha = 0.5)
plt.plot(linear,y2,':xb',alpha = 0.5)
```

```python
x = plt.gca().xaxis
```
