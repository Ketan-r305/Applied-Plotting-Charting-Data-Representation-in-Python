```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
x1 = np.random.normal(-2.5, 1, 10000)
x2 = np.random.gamma(2, 1.5, 10000)
x3 = np.random.exponential(2, 10000)+7
x4 = np.random.uniform(14,20, 10000)
```

```python
plt.figure()
plt.subplot(221)
plt.hist(x1,bins = 20,density = True,color = 'r')
plt.subplot(222)
plt.hist(x2,bins = 20,normed = True,color = 'm')
plt.subplot(223)
plt.hist(x3,bins = 20,normed = True,color = 'y')
plt.subplot(224)
plt.hist(x4,bins = 20,normed = True,color = 'k')
plt.show()
```
